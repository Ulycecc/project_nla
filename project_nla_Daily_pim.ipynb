```python
data = pd.read_csv("Daily-train.csv")
```

```python
def make_features(data, lag):
    n = data.shape[0]
    A = np.zeros([n-lag, lag])
    for i in range(lag):
        A[:, lag-i-1] = data[i:n-lag+i]
    return A


def pim(X):
    return np.dot(np.linalg.inv(np.dot(X.T, X)), X.T)
```

```python
def compary(data, lag, iter):
    rez = []
    rez1 = []

    for i in range(1, iter):
        train, test = train_test_split(
            data, shuffle=False, test_size=i*.95/iter)

        A_train = make_features(train, lag)
        b_train = train[lag:]
        A_test = make_features(test, lag)
        b_test = test[lag:]

        Q, R = np.linalg.qr(A_train)
        x = np.linalg.inv(R).T @ Q.T @ b_train
        y_qr = A_test@x
        mse = np.var((b_test - y_qr), ddof=1)
        rez.append(mse**0.5)

        x_pim = pim(A_train)@b_train
        y_pim = A_test@x_pim
        mse = np.var((b_test - y_pim), ddof=1)
        rez1.append(mse**0.5)

    # d = {'QR': rez, 'ПОМ': rez1}
    # df = pd.DataFrame(data=d)
    # df.plot(figsize=(20, 6))

    # d_n = {'QR': rez_n, 'ПОМ': rez1_n}
    # df_n = pd.DataFrame(data=d_n)
    # df_n.plot(figsize=(20, 6))
    # plt.show()

    return np.mean(rez)-np.mean(rez1)
```
